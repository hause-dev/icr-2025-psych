```python
for i in range(0, 945):
    if arr[i, 6] == str("Anxiety disorder") and arr[i, 5] != "NA" and arr[i, 4] != "NA":
        row = [arr[i, 2], arr[i, 4], arr[i, 5]]  
        for mood in range(8, 122):
            row.append(arr[i, mood])
        row.append("1")
        total_data.append(row)

for i in range(0, 945):
    if arr[i, 6] == str("Healthy control") and arr[i, 5] != "NA" and arr[i, 4] != "NA":
        row = [arr[i, 2], arr[i, 4], arr[i, 5]]  
        for hc in range(8, 122):
            row.append(arr[i, hc])
        row.append("0")
        total_data.append(row)
print(len(total_data[0]), total_data[0])
print(len(final_columns), final_columns)
# Build the DataFrame
df = pd.DataFrame(total_data, columns=final_columns)
iq_col = final_columns[2]
edu_col = final_columns[1]
for col in [edu_col, iq_col]:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.upper()
        .replace("NA", pd.NA)
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df = df.dropna(subset=[col])
#np.save("/home/novo/icr-2025-psych/AEIaddictivehc_allPSD.npy")

# Save to CSV
df.to_csv("/home/novo/icr-2025-psych/AEIanxietyhc_allPSD.csv", index=False)
```
